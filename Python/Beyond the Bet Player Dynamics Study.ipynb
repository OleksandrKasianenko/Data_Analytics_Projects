```python
daily = df.groupby('date', as_index=False)['bet_sum'].sum().sort_values('date')

mean = daily['bet_sum'].mean()
std  = daily['bet_sum'].std()
thr  = 3*std
outs = daily[(daily['bet_sum'] > mean + thr) | (daily['bet_sum'] < mean - thr)]

import matplotlib.pyplot as plt
plt.close('all')                       # прибрати старі фігури
plt.figure(figsize=(16,6))
plt.plot(daily['date'], daily['bet_sum'], label='Bet sum')
plt.scatter(outs['date'], outs['bet_sum'], label='Outliers')
plt.axhline(mean + thr, linestyle='--', label='Threshold +')
plt.axhline(mean - thr, linestyle='--', label='Threshold -')
plt.title(f'Bet Sum with Outliers (n={len(outs)})')
plt.xlabel('Date'); plt.ylabel('Bet Sum'); plt.legend(); plt.tight_layout(); plt.show()

```

At the end of the month, players were playing more rounds but betting less per round — leading to a drop in the overall bet_sum.

The most significant factor influencing results is system_id, followed by time trend and game.
